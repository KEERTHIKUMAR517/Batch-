```python
retail['Country'].value_counts()
```

```python
france_data = retail[retail['Country'] == 'France']
```

### Data Preparation

```python
france_data_pivot_table = pd.pivot_table(data= france_data,values='Quantity',index='InvoiceNo',columns='Description').fillna(0)
```

```python
france_encoded = france_data_pivot_table.applymap(lambda x:1 if x > 0 else 0)
```

# Model Building

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_itemsets = apriori(france_encoded,min_support=0.02,use_colnames=True)
```

```python
freq_itemsets.to_csv('freq.csv')
```

```python
best_associates = association_rules(freq_itemsets,metric='confidence',min_threshold=0.6,)
best_associates
```

```python
best_associates.to_csv('best_associates.csv')
```
