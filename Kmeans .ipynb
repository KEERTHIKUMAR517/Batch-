```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
del iris['species']
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
X = minmax.fit_transform(iris)
X
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)

fit = kmeans.fit(X)
y_pred = kmeans.predict(X)
```

```python
import pandas as pd
target = pd.DataFrame(y_pred,columns=['target'])
```

```python
iris['target'] = target
iris
```

```python
iris[iris['target']==0]
```

```python
iris[iris['target']==1]
```

```python
X = pd.DataFrame(X)
```

```python
wcss =[]

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) 
```

```python
wcss
```

```python

plt.plot(range(1,11),wcss)
```
